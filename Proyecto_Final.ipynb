```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Entrenamiento
dt_classifier = DecisionTreeClassifier(random_state=1)
dt_classifier.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred_dt = dt_classifier.predict(X_test)

# Matriz de confusión
cm_dt = confusion_matrix(y_test, y_pred_dt)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Blues',
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Decision Tree Confusion Matrix')
plt.tight_layout()
plt.show()

print("=== Decision Tree ===")
print(classification_report(y_test, y_pred_dt, target_names=label_encoder.classes_))

# Gráfica de aprendizaje
train_sizes, train_scores, test_scores = learning_curve(
    dt_classifier, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), random_state=42
)

train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes.astype(int), train_scores_mean, 'o-', color='blue', label='Training score')
plt.plot(train_sizes.astype(int), test_scores_mean, 'o-', color='green', label='Cross-validation score')
plt.title('Decision Tree Learning Curve')
plt.xlabel('Training examples')
plt.ylabel('Accuracy')
plt.legend(loc='best')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
import matplotlib.pyplot as plt
import numpy as np

# Para la curva ROC y el AUC promedio, necesitamos las probabilidades predichas
# Para DecisionTreeClassifier, podemos usar predict_proba
y_score_dt = dt_classifier.predict_proba(X_test)

# Binarizar las etiquetas para la curva ROC (One-vs-Rest)
y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
n_classes = y_test_bin.shape[1]

# Calcular la curva ROC y el AUC para cada clase
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_dt[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Calcular la micro-media ROC y AUC (promedio de todos los pares verdadero-falso positivos)
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_score_dt.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Calcular la macro-media ROC y AUC (promedio de AUC de cada clase)
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

# Interpolar todas las curvas ROC en este conjunto de puntos
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

# Promediar y calcular el AUC
mean_tpr /= n_classes

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])


# Graficar la curva ROC promedio (Macro-media)
plt.figure(figsize=(8, 6))
plt.plot(fpr["macro"], tpr["macro"],
         label=f'Macro-promedio Curva ROC (área = {roc_auc["macro"]:.2f})',
         color='deeppink', linestyle=':', linewidth=4)

# Graficar la curva ROC promedio (Micro-media)
plt.plot(fpr["micro"], tpr["micro"],
         label=f'Micro-promedio Curva ROC (área = {roc_auc["micro"]:.2f})',
         color='navy', linestyle=':', linewidth=4)

# # Graficar la curva ROC para cada clase (opcional, descomentar si se desea ver)
# colors = plt.cm.get_cmap('tab10', n_classes)
# for i in range(n_classes):
#     plt.plot(fpr[i], tpr[i], color=colors(i), lw=2,
#              label=f'Curva ROC clase {label_encoder.classes_[i]} (área = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC Promedio - Árbol de Decisión')
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.show()

print("\n=== Área bajo la Curva ROC (AUC) ===")
print(f"AUC Macro-promedio: {roc_auc['macro']:.2f}")
print(f"AUC Micro-promedio: {roc_auc['micro']:.2f}")
```

```python
import numpy as np
from scipy.spatial.distance import cdist

# Calcular centroides por actividad
centroids = pca_df.groupby('actividad')[['PC1', 'PC2']].mean()

# Calcular distancias euclidianas entre centroides
dist_matrix = cdist(centroids.values, centroids.values, metric='euclidean')

# Crear DataFrame para facilitar lectura
dist_df = pd.DataFrame(dist_matrix, index=centroids.index, columns=centroids.index)

print("Distancias entre centroides de actividades (espacio PCA):")
print(dist_df)

# Opcional: mostrar las parejas con menor distancia (más cercanas)
dist_pairs = []

for i, act1 in enumerate(centroids.index):
    for j, act2 in enumerate(centroids.index):
        if i < j:
            dist_pairs.append((act1, act2, dist_matrix[i, j]))

# Ordenar por distancia ascendente
dist_pairs_sorted = sorted(dist_pairs, key=lambda x: x[2])

print("\nParejas de actividades más cercanas:")
for act1, act2, dist in dist_pairs_sorted[:5]:  # Mostrar top 5 parejas más cercanas
    print(f"{act1} - {act2}: distancia = {dist:.4f}")
```

# Matriz de correlacion

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from scipy.stats import spearmanr

# Copia del DataFrame original y codificación
df_temp = df.copy()
df_temp['actividad_cod'] = LabelEncoder().fit_transform(df_temp['actividad'])

# Extraer nombres de actividades
activity_names = df_temp['actividad'].unique()

# Crear matriz vacía
correlation_matrix = pd.DataFrame(index=activity_names, columns=activity_names)

# Calcular correlación entre actividades usando muestras individuales
for act1 in activity_names:
    for act2 in activity_names:
        vals1 = df_temp[df_temp['actividad'] == act1].drop(columns=['persona', 'repeticion', 'actividad', 'actividad_cod']).values.flatten()
        vals2 = df_temp[df_temp['actividad'] == act2].drop(columns=['persona', 'repeticion', 'actividad', 'actividad_cod']).values.flatten()
        corr, _ = spearmanr(vals1, vals2)
        correlation_matrix.loc[act1, act2] = corr

# Convertir a float para mostrar valores numéricos reales
correlation_matrix = correlation_matrix.astype(float)

# Mostrar la matriz en consola
print("Matriz de correlación entre actividades (Spearman):")
print(correlation_matrix.round(3))  # redondear para mejor lectura
```
